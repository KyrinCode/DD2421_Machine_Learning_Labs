## Assignment 2:
Explain entropy for a uniform distribution and a non-uniform distribution, present some example distributions with high and low entropy.

## Answer:
Uniform distribution is a circumstance that every result has the same probability $p$, just like tossing a coin. Assuming uniform distribution has $n$ kinds of results, the entropy will be $-\frac{1}{n}log_2\frac{1}{n}$, since $p=\frac{1}{n}$.

Non-uniform distribution's results have different probability, just like weather. Assuming there are $n$ results, $i$th result has probability $p_i$. The entropy will be $\sum-p_ilog_2p_i$.

## Assignment 6:
Explain pruning from a bias variance trade-off perspective.

## Answer:
Given a training dataset, if we design an algorithm that is too complex, then the training dataset will be fitted very well and the bias can be very low. Constructing a decision tree without pruning may be this kind of circumstance. the  model can fit the training data so well, but when it is fed with new data it will lack of the ability of generalization, namely low variance or overfitting. Pruning is a method to lower the complex of our decision tree, so that the bias is improved but the variance is lowered, which is the meaning of bias-variance trade-off. Then the model will be more flexible and generalize well.


## Assignment 7:
Evaluate the effect pruning has on the test error for the monk1 and monk3 datasets, in particular determine the optimal partition into training and pruning by optimizing the parameter fraction. Plot the classification error on the test sets as a function of the parameter fraction âˆˆ {0.3, 0.4, 0.5, 0.6, 0.7, 0.8}.

Note that the split of the data is random. We therefore need to compute the statistics over several runs of the split to be able to draw any conclusions. Reasonable statistics includes mean and a measure of the spread. Do remember to print axes labels, legends and data points as you will not pass without them.

```python
fraction_list = [0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 0.95, 0.98]
times = 100

def partition(data, fraction):
    ldata = list(data)
    random.shuffle(ldata)
    breakPoint = int(len(ldata) * fraction)
    return ldata[:breakPoint], ldata[breakPoint:]

def prune(dataset, test_set):
    prune_result = []
    for fraction in fraction_list:
        training_set, validation_set = partition(dataset, fraction)
        tree = d.buildTree(training_set, m.attributes)
        forest = d.allPruned(tree)
        best_value = d.check(tree, validation_set)
        temp_tree = 0
        best_tree = tree
        for t in forest:
            temp_value = d.check(t, validation_set)
            if best_value < temp_value:
                best_value = temp_value
                best_tree = t
        prune_result.append(1 - d.check(best_tree, test_set))
    return prune_result

monk1_pruned_error = []
monk3_pruned_error = []
for i in range(times):
    monk1_pruned_error.append(prune(m.monk1, m.monk1test))
    monk3_pruned_error.append(prune(m.monk3, m.monk3test))

monk1_pruned_error = np.transpose(monk1_pruned_error)
monk3_pruned_error = np.transpose(monk3_pruned_error)

monk1_pruned_error_mean = np.mean(monk1_pruned_error, axis=1)
monk3_pruned_error_mean = np.mean(monk3_pruned_error, axis=1)
monk1_pruned_error_std = np.std(monk1_pruned_error, axis=1)
monk3_pruned_error_std = np.std(monk3_pruned_error, axis=1)

plt.plot(fraction_list, monk1_pruned_error_mean, color='red', marker='o', label="Means")
plt.title("Means of Error vs Fractions on MONK-1")
plt.xlabel("Fractions")
plt.ylabel("Means of Error")
plt.legend(loc='upper right', frameon=False)
plt.show()

for i in range(0, len(fraction_list)):
    print([fraction_list[i], monk1_pruned_error_mean[i]])

plt.plot(fraction_list, monk3_pruned_error_mean, color='blue', marker='o', label="Means")
plt.title("Means of Error vs Fractions on MONK-3")
plt.xlabel("Fractions")
plt.ylabel("Means of Error")
plt.legend(loc='upper right', frameon=False)
plt.show()

for i in range(0, len(fraction_list)):
    print([fraction_list[i], monk3_pruned_error_mean[i]])
```

From the diagram, we can tell that the lowest means of error is when the fraction is nearby 0.8. It's easy to understand that the larger training set is, the model can be trained better. But if we research more into this, we can find that when the fraction continues growing, the mean of error may grows up too. So it's not the best choice to let the fraction of training set and validation set be as big as possible. The samples of validation set should be enough. According to the diagram, it's better to let the fraction be 0.8.
