# Support Vector Machines

## Import packages

```python
import numpy, random, math
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import random
```

## Generate test dataset.

```python
numpy.random.seed(100)
classA = numpy.concatenate((
    numpy.random.randn(10, 2) * 0.2 + [1.5, 0.5],
    numpy.random.randn(10, 2) * 0.2 + [-1.5, 0.5],
    ))
classB=numpy.random.randn(20, 2) * 0.2 + [0.0, -0.5]

inputs=numpy.concatenate((classA, classB))
targets=numpy.concatenate((numpy.ones(classA.shape[0]), -numpy.ones(classB.shape[0])))

N = inputs.shape[0] # Number of rows (samples)

permute = list(range(N))
random.shuffle(permute)
inputs = inputs[permute,:]
targets = targets[permute]
```

## Design kernels.

```python
#linear kernel
def linear_kernel(x, y):
    return numpy.dot(x, y)
    
# polynomial kernels
def polynomial_kernel(x, y):
    p = 2
    return numpy.power(numpy.dot(x, y) + 1, p)

# radial basis function (RBF) kernels
def RBF_kernel(x, y):
    # the parameter sigma is used to control the smoothness of the boundary
    sigma = 0.1
    return math.exp(-math.pow(numpy.linalg.norm(numpy.subtract(x, y)), 2)/(2 * math.pow(sigma,2)))

kernel = RBF_kernel
```

## Initialize PMatrix once for all.

```python
PMatrix = numpy.zeros((N,N))
for i in range(N):
    for j in range(N):
        PMatrix[i][j] = targets[i] * targets[j] * kernel(inputs[i], inputs[j])
```

```python
def zerofun(alpha):
    return numpy.dot(alpha, targets)
```

```python
def objective(alpha):
    return (1/2) * numpy.dot(alpha, numpy.dot(alpha, PMatrix)) - numpy.sum(alpha)
```

## Minimize vector $a$

```python
start = numpy.zeros(N)
C = 10

# If have an upper constraint
# B = [(0, C) for b in range(N)]

# If only have a lower bound:
B = [(0, None) for b in range(N)]

# XC is the second half of equation 10; given as a dictionary with fields type and fun
# In the following example, zerofun is a function which calculates the value which
# ...should be constrained to zero.
XC = {'type':'eq', 'fun':zerofun}

ret = minimize(objective, start, bounds=B, constraints=XC)
if (not ret['success']):
    raise ValueError('Cannot find optimizing solution')
```

## Extract nonzero $\alpha$

```python
alpha = ret['x']
nonzero = [(alpha[i], inputs[i], targets[i]) for i in range(N) if abs(alpha[i]) > 10e-5]
```

## Calculate the b value using equation (7).

```python
def b_value():
    b_sum = 0
    for v in nonzero:
        b_sum += v[0] * v[2] * kernel(v[1], nonzero[0][1])
    return b_sum - nonzero[0][2]
```

## Implement the indicator function using equation (6)

```python
def indicator(x, y, b):
    ind_sum = 0
    for v in nonzero:
        ind_sum += v[0] * v[2] * kernel([x, y], v[1])
    return ind_sum - b
```

## Plot

```python
b = b_value()

# Plot data points
plt.plot([p[0] for p in classA], [p[1] for p in classA], 'b.')
plt.plot([p[0] for p in classB], [p[1] for p in classB], 'r.')
plt.plot([p[1][0] for p in nonzero], [p[1][1] for p in nonzero], 'g.')

# Plot contour
xgrid = numpy.linspace(-5, 5)
ygrid = numpy.linspace(-4, 4)
grid = numpy.array([[indicator(x, y, b) for x in xgrid] for y in ygrid])
plt.contour(xgrid, ygrid, grid, (-1.0,0.0,1.0), colors=('red','black','blue'), linewidths=(1,3,1))

# plt.savefig('figures/svmplot.pdf') # Save a copy in a file
plt.show() # Show the plot on the screen
```

## Q3:

The non-linear kernels have parameters; explore how they influence the decision boundary. Reason about this in terms of the bias variance trade-off.

## Answer:

Polynomial kernels - p:

When the parameter p increases, the complexity will also increase. It's obvious that bias will decrease but variance will increase when parameter p is higher. So, the order of the polynomial p should be appropriate to get the best model.

RBF kernels - $\sigma$:

When $\sigma$ increases, the function's smoothness will increase. So a higher sigma will tend to fit a more straight line. The complexity decreases which bring about decrease of variance and increase of bias. So, the parameter $\sigma$ should be appropriate to get the best model according to different datasets.

## Q5:

Imagine that you are given data that is not easily separable. When should you opt for more slack rather than going for a more complex model (kernel) and vice versa?

## Answer:

Slack is used in the case that there are some noise points that interfere the training process. If we can tell that the complexity is already enough for the dataset, then we should consider that we should introduce slack to tolerate noise points.

In reverse, if we know that the dataset is of high quality but the performance is still not good enough, then it should be that the complexity of model is not enough. So, we need to introduce a more complex model.
